# 1. Set user-agent to prevent being blocked

# 2. Find the information of the first store in page one

```python
from bs4 import BeautifulSoup
import requests
#get url from Michelin website(location:Taipei)
page = requests.get("https://guide.michelin.com/tw/zh_TW/taipei-region/taipei/restaurants?lat=24.1506212&lon=120.6433008")
#use soup to retreive html content
soup = BeautifulSoup(page.content, 'html.parser')
#class"col-lg-12 search-results__column" contains 20 store
rest = soup.find(class_="col-lg-12 search-results__column")
#each store is under the subdirectory of 
#class"col-lg-12 search-results__column", it contains 20 stores,
#each store is assigned under class"col-md-6 col-lg-6 col-xl-3"
items = rest.find_all(class_="col-md-6 col-lg-6 col-xl-3")
#we first take out the first store in the items list
rest1 = items[1]
print(rest1.prettify())
```

# 3. Retreive data from the first store

```python
restname = rest1.find(class_="card__menu-content--title last").get_text().strip() #store name
restclass = rest1.find(class_="card__menu-footer--price").get_text().strip() #category
imagesurl1 = rest1.findAll('img', alt='') #image url 1
imagesurl2 = rest1.find('img')['src'].strip() #image url 2
print(restname)
print(restclass)
print(imagesurl1)
print(imagesurl2)
#print out in dataframe
import pandas as pd
storeOne = pd.DataFrame({
    "restname": restname,
    "restclass": restclass,
    "imagesURL1": imagesurl1,
    "imagesURL2":imagesurl2
})
storeOne
```

# 4. Retrieve picture url from 20 stores

```python
rest = soup.find(class_="col-lg-12 search-results__column")
items = rest.find_all(class_="col-md-6 col-lg-6 col-xl-3") 
length = len(items) #count stores
picurl = [] #create null list
picurlTable = []
#loop over 20 stores
for i in range(length):
    #append all items to the list
    picurl.append(items[i])
    #append stores to table list and find image url from 20 stores
    picurlTable.append(picurl[i].find('img')['src'].strip())
    #find image url from 20 stores
    print(picurl[i].find('img')['src'].strip())
#print out in dataframe
import pandas as pd
#present full image url
pd.set_option('display.max_colwidth', -1)
picURL = pd.DataFrame({
    "picURL":picurlTable
})
picURL
```

# 5. Retrieve store url from 20 stores

```python
rooturl = 'https://guide.michelin.com'
link = soup.findAll("a", {"class": "link"})
leng = len(link) #count links
linkList = []
sitelinkList = []
for x in range(leng):
    linkList.append(link[x])
    itemurl = linkList[x]['href']
    #since the url we retreive only contain the url of the store,
    #we should add the rooturl to get a complete url 
    siteurl = rooturl + itemurl
    sitelinkList.append(siteurl)
    print(siteurl)
#print out in dataframe
siteURL = pd.DataFrame({
    "siteURL":sitelinkList
})
siteURL
```

# 6. Retrieve store name from 20 stores

```python
link = soup.findAll("a", {"class": "link"})
leng = len(link) #count links
storeName = []
storeNamelist = []
for x in range(leng):
    storeName.append(link[x])
    #store name is contained in the url-"a arial-label" 
    print(storeName[x]['aria-label'][5:])#truncate first five words(open )
    storeNamelist.append(storeName[x]['aria-label'][5:])
#print out in dataframe
storeName = pd.DataFrame({
    "storeName":storeNamelist
})
storeName
```
